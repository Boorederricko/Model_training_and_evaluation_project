```python
import cv2
import os
import time
```

```python
# Define the names of the classes and directories to save images
classes = ['Clear_path']
base_dir = 'C:\\Users\\hp\\TensorFlowlite\\Main_dataset'
```

```python
# Initialize the webcam
cap = cv2.VideoCapture(0)
```

```python
# Set the desired capture rate (10 milliseconds per image)
capture_rate = 5  # milliseconds
```

```python
# Initialize counters for images collected 
image_counts = [0] * len(classes)

# Add a 5-second delay before capturing images
print("Waiting for 5 seconds...")
time.sleep(5)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imshow('Collecting Images', frame)

    # Introduce a delay to achieve the desired capture rate
    time.sleep(capture_rate)

    # Check if it's time to capture a new image
    current_time = time.time()
    for class_idx, class_name in enumerate(classes):
        if (current_time - image_counts[class_idx] * capture_rate) >= capture_rate and (image_counts[class_idx] < 100):
            image_filename = os.path.join(base_dir, class_name, f'{class_name}_{image_counts[class_idx] + 1}.jpg')
            cv2.imwrite(image_filename, frame)
            print(f'Saved: {image_filename}')
            image_counts[class_idx] += 1

    # Check if (x) images have been collected 
    if all(count == 200 for count in image_counts):
        break

    # Check for user input to quit the program
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break

# Release the webcam and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
